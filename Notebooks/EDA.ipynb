# EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.set_option('display.max_columns', 300)
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv('data/california_wildfires.csv')
```

```python
df.head()
```

```python
sns.countplot(df.fire_started)
```

```python
year_ex = df.loc[df['date'] < '2014-01-01']
year_ex['date'] = pd.to_datetime(year_ex['date'])
year_ex = year_ex.resample('M', on = 'date').mean()
year_ex = year_ex.reset_index()
```

```python
year_ex
```

```python
plt.plot(year_ex['date'], year_ex.fire_started);
```

```python
df['month'] = pd.DatetimeIndex(df.date).month
```

```python
month_gpby = df.groupby('month')['fire_started'].sum()
```

```python
month_gpby
```

```python
month_gpby.plot(kind = 'bar')
```

The majority of fires occur between the months of May and September.

```python
year_gpby = df.groupby('year')['fire_started'].sum()
```

```python
year_gpby.plot(kind = 'bar')
```

On average there is an increase in wildfires over the years.

```python
county_gpby = df.groupby('county')['fire_started'].sum()
```

```python
county_gpby.plot(kind = 'barh', figsize = (10, 16))
```

There are certain counties that are more likely to have fires than others. The number one county for wildfires is Riverside.
