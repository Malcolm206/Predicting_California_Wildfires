```python
#data_list = []
for n in range(len(websites)):
    driver.get(websites[n])
    timeout = 5
    try:
        element_present = EC.presence_of_element_located((By.ID, 'main'))
        WebDriverWait(driver, timeout).until(element_present)
    except TimeoutException:
        print("Timed out waiting for page to load")
    finally:
        print("Page loaded")
    time.sleep(5)
    weather_data = pd.DataFrame()
    for i in range(7):
        click_view = driver.find_element(By.XPATH, '//*[@id="inner-content"]/div[2]/div[1]/div[1]/div[1]/div/lib-date-selector/div/input').click()
        time.sleep(5)
        for x in range(12):
            click_month = driver.find_element(By.XPATH, '//*[@id="inner-content"]/div[2]/div[1]/div[1]/div[1]/div/lib-date-selector/div/select[1]').click()
            click_choice = driver.find_element(By.XPATH, month_x_path[x]).click()
            click_view = driver.find_element(By.XPATH, '//*[@id="inner-content"]/div[2]/div[1]/div[1]/div[1]/div/lib-date-selector/div/input').click()
            timeout = 5
            try:
                element_present = EC.presence_of_element_located((By.ID, 'main'))
                WebDriverWait(driver, timeout).until(element_present)
            except TimeoutException:
                print("Timed out waiting for page to load")
            finally:
                print("Page loaded")
            time.sleep(5)
            driver.page_source
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            table = soup.find_all('table')
            df = pd.read_html(str(table))[-7:]
            result = pd.concat(df, axis = 1)
            result.drop(0, axis = 0, inplace = True)
            weather_data = weather_data.append(result)
        click_year = driver.find_element(By.XPATH, '//*[@id="inner-content"]/div[2]/div[1]/div[1]/div[1]/div/lib-date-selector/div/select[2]')
        click_choice_year = driver.find_element(By.XPATH, year_x_path[i]).click()
    weather_data.columns = ['Date', 'Max Air Temp (F)', 'Avg Air Temp (F)', 'Min Air Temp (F)', 
                            'Max Dew Point (F)', 'Avg Dew Point (F)', 'Min Dew Point (F)', 
                            'Max Rel Hum (%)', 'Avg Rel Hum (%)', 'Min Rel Hum (%)', 
                            'Max Wind Speed (mph)', 'Avg Wind Speed (mph)', 'Min Wind Speed (mph)', 
                            'Max Pressure (Hg)', 'Avg Pressure (Hg)', 'Min Pressure (Hg)', 'Precip (in)']
    weather_data.drop(['Max Dew Point (F)', 'Min Dew Point (F)', 'Max Wind Speed (mph)',
                        'Min Wind Speed (mph)', 'Max Pressure (Hg)', 'Min Pressure (Hg)'],
                        axis = 1)
    data_list.append(weather_data)
```
